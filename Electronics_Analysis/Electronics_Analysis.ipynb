```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in Num_Columns:
    plt.figure()
    plt.tight_layout()
    sns.set(rc={"figure.figsize":(8, 5)})
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True)
    plt.gca().set(xlabel= i,ylabel='Frequency')
    sns.boxplot(data[i], ax=ax_box , linewidth= 1.0)
    sns.histplot(data[i], ax=ax_hist , bins = 10,kde=True)
```

```python
for i in Num_Columns:
    Q1=data.describe().loc["25%",i]
    Q3=data.describe().loc["75%",i]
    IQR=Q3-Q1
    b = (Q3+IQR*1.5)
    c = (Q1-IQR*1.5)
    data[i].mask((data[i]>b),data[i].mean(),inplace =True)
    data[i].mask((data[i]<c),data[i].mean(),inplace =True)
```
